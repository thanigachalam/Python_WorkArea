```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(color_codes=True)
%matplotlib inline
```

```python
auto = pd.read_csv("Automobile.csv")
```

```python
auto.head()
```

```python
#3types
#univariate distribute
```

```python
sns.distplot(auto["normalized_losses"])
plt.show()
```

```python
sns.distplot(auto["fuel_type"])
plt.sow()

#error due to category variable in distrubtion chart
```

```python
sns.distplot(auto['city_mpg'])
plt.show()
```

```python
sns.distplot(auto['city_mpg'],kde=False,rug=True)
plt.show()

#rug true will show to sequence or frequency lines below
#we can find the dense and hight based on the value
```

```python
#byvariate analysis.

sns.jointplot(auto['engine_size'],auto['horsepower'])

plt.show()
```

```python
sns.jointplot(auto['engine_size'], auto['horsepower'], kind="hex")
#kind : { "scatter" | "reg" | "resid" | "kde" | "hex" },
plt.show()
```

```python
sns.jointplot(auto['engine_size'], auto['horsepower'], kind="reg")
#kind : { "scatter" | "reg" | "resid" | "kde" | "hex" },
plt.show()
```

```python
sns.jointplot(auto['engine_size'], auto['horsepower'], kind="kde")
```

```python
sns.pairplot(auto[['normalized_losses','engine_size','horsepower']])
#dignal histogram and 
```

```python
sns.stripplot(auto['fuel_type'], auto['horsepower'])
```

```python
#adding jitter

sns.stripplot(auto['fuel_type'], auto['horsepower'], jitter=True)
```

```python
sns.swarmplot(auto['fuel_type'],auto['horsepower'])
#ordered set of dots
```

```python
sns.boxplot(auto['number_of_doors'], auto['horsepower'], hue=auto['fuel_type'])
```

```python
sns.barplot(auto['body_style'],auto['horsepower'])
```

```python
sns.barplot(auto['body_style'],auto['horsepower'],hue=auto['engine_location'])
```

```python
sns.countplot(auto['body_style'])
```

```python
sns.countplot(auto['body_style'],hue=auto['engine_location'])
```

```python
sns.pointplot(auto['fuel_system'],auto['horsepower'],hue=auto['number_of_doors'])
```

```python
#factor plot

sns.factorplot(x="fuel_type",
              y="horsepower",
              hue="number_of_doors",
              col="engine_location",
              data=auto,
              kind="swarm")
```

```python
#{``point``, ``bar``, ``count``, ``box``, ``violin``, ``strip``}

#count willl have only x axis, only need single axis value

sns.factorplot(x="fuel_type",
              y="horsepower",
              hue="number_of_doors",
              col="engine_location",
              data=auto,
              kind="box")
```

```python
sns.lmplot(x="horsepower",y="peak_rpm",data=auto,hue="fuel_type")

#confidance interval expanding
```
