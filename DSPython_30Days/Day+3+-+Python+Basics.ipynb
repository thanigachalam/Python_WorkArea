```python
class FileHandler:
    
    def writeData(self,content):
        print("you are in write Data", content)
        fo = open(r"D:\\DW\\Masters program in Data Science_LJMU\\Python_Live_Class\createdfromcode.txt","w")
        fo.write(content)
        fo.close()
        print("Hurry! Data saved!")
        
    def readData(self,path):
        print("you are in readData",path)
        fo = open(path,"r")
        for line in fo:
            print(line)
        fo.close()

class Program(FileHandler): #observe the inheritance syntax
    
    def AcceptInput(self):
        data = input("Enter input data to be saved!")
        self.writeData(data) #oberve calling function of base class

p = Program()
#p.AcceptInput()
p.readData(r"D:\\DW\\Masters program in Data Science_LJMU\\Python_Live_Class\createdfromcode.txt")
```
