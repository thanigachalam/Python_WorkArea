# Deal missing data and group-by options

```python
dataframe = pd.DataFrame({'Cricket':[1,2,np.nan,4,6,7,2,np.nan],
                         'Baseball':[5,np.nan,np.nan,5,7,2,4,5],
                         'Tennis':[1,2,3,4,5,6,7,8]})
```

```python
dataframe
```

```python
dataframe.dropna()
#droping na rows
```

```python
#droping na columns
dataframe.dropna(axis=1)
```

```python
dataframe.dropna(thresh=2)
#if the valid input of 2 then that row will be shown hence others will get deleted
```

```python
#replacing na values with 0
dataframe.fillna(value=0)
```

```python
#replaing na value with mean value
dataframe['Baseball'].fillna(value=dataframe['Baseball'].mean())
```

```python
dataframe['Baseball'].mean()
```

# Merging similar to SQL logic

```python
dafa1 = pd.DataFrame({'CustID':['101','102','103','104'],
                     'Sales':[13456,45321,54385,53212],
                     'Priority':['CAT0','CAT1','CAT2','CAT3'],
                     'Prime':['yes','no','no','yes']},
                    index=[0,1,2,3])
```

```python
dafa2 = pd.DataFrame({'CustID':['101','103','104','105'],
                     'Sales':[13456,54385,53212,4534],
                     'Payback':['CAT4','CAT5','CAT6','CAT7'],
                     'Imp':['yes','no','no','no']},
                     index=[4,5,6,7])
```

```python
pd.concat([dafa1,dafa2])
```

```python
pd.concat([dafa1,dafa2],axis=1)
```

```python
pd.merge(dafa1,dafa2,how='outer',on='CustID')
```

```python
dafa3 = pd.DataFrame({'Q1': ['101','102','103'],
                     'Q2': ['201','202','203']},
                    index=['10','11','12'])
```

```python
dafa4 = pd.DataFrame({'Q3': ['301','302','303'],
                     'Q4': ['401','402','403']},
                    index=['10','12','13'])
```

```python
dafa3.join(dafa4)
```

```python
dafa3.join(dafa4, how='left')
```

```python
dafa3.join(dafa4, how='right')
```
